```python
school = pd.read_csv('datasets/schoolimprovement2010grants.csv')
school.head()
```

```python
df_1 = pd.read_csv('datasets/insurance_premiums.csv')
df_1.head()
```

```python
g = sns.PairGrid(df_1, vars=['fatal_collisions', 'premiums'])
g2 = g.map(sns.scatterplot)

plt.show()
```

```python
g = sns.PairGrid(df_1, vars=['fatal_collisions', 'premiums'])
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)

plt.show()
```

```python
sns.pairplot(df_1, vars=['fatal_collisions', 'premiums'], kind='reg', diag_kind='hist')

plt.show()
```

```python
bike.head()
```

```python
g = sns.JointGrid(
    x='hum',
    y='total_rentals',
    data=bike,
    xlim=(0.1, 1.0)
)

g.plot(sns.regplot, sns.histplot)

plt.show()
```

```python
g = sns.JointGrid(
    x='hum',
    y='total_rentals',
    data=bike,
    xlim=(0.1, 1.0)
)

g.plot_joint(sns.kdeplot)
g.plot_marginals(sns.kdeplot, fill=True)

plt.show()
```
