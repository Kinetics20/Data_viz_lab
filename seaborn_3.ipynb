```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/schoolimprovement2010grants.csv')
df.head()
```

```python
sns.displot(df['Award_Amount'],  kind='kde', fill=True, rug=True)
plt.show()
```

```python
df = pd.read_csv('datasets/insurance_premiums.csv')
df.head()
```

```python
sns.regplot(x='insurance_losses', y='premiums', data=df)
plt.show()
```

```python
sns.lmplot(x='insurance_losses', y='premiums', data=df, col='Region')
plt.show()
```
