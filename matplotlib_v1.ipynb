```python
austin_weather['MONTH'] = pd.to_datetime(austin_weather['DATE'], format='%m').dt.strftime('%b')
austin_weather['MONTH'].head(12)
```

```python
months = {
    1: "January",
    2: "February",
    3: "March",
    4: "April",
    5: "May",
    6: "June",
    7: "July",
    8: "August",
    9: "September",
    10: "October",
    11: "November",
    12: "December"
}
seatle_weather['MONTH'] = seatle_weather['DATE'].apply(lambda x: months[x])
```

```python
# austin prcp vs seattle prcp (MLY-PRCP-25PCTL, MLY-PRCP-75PCTL)

austin = austin_weather.groupby('MONTH', sort=False)[['MLY-PRCP-NORMAL', 'MLY-PRCP-25PCTL', 'MLY-PRCP-75PCTL']].mean()
seattle = seatle_weather.groupby('MONTH', sort=False)[['MLY-PRCP-NORMAL', 'MLY-PRCP-25PCTL', 'MLY-PRCP-75PCTL']].mean()
```

```python
def annotate_with_marker(df, ax, title='Max value',loc=(0, 0),**arrowprops):
    max_idx_, max_val_ = df['MLY-PRCP-NORMAL'].idxmax(), df['MLY-PRCP-NORMAL'].max()

    props = dict(arrowstyle='->', color='m') | arrowprops
    ax.scatter(max_idx_, max_val_, color='orange', marker='o')
    ax.annotate(title, 
               xy=(max_idx_, max_val_),
               xytext=(max_idx_ , max_val_ + 5),
               arrowprops=dict(arrowstyle='->', color='m')
              )
    
```

```python
fig, ax = plt.subplots(2, 1, sharey=True)

ax[0].plot(austin.index, austin['MLY-PRCP-NORMAL'],color='b')
ax[0].plot(austin.index, austin['MLY-PRCP-25PCTL'], linestyle='--', color='b')
ax[0].plot(austin.index, austin['MLY-PRCP-75PCTL'], linestyle='--', color='b')

# austin_max_idx, austin_max_prcp = austin['MLY-PRCP-NORMAL'].idxmax(), austin['MLY-PRCP-NORMAL'].max()

ax[0].axvline('Jun', color='steelblue', linestyle='--', linewidth=0.3)
ax[0].axhline(austin['MLY-PRCP-NORMAL'].max(), color='steelblue', linestyle='--', linewidth=0.3)
annotate_with_marker(austin, ax[0], loc=(0, 5))

# ax[0].scatter(austin_max_idx, austin_max_prcp, color='orange', marker='o')
# ax[0].annotate('Max avg month percipitation', 
#                xy=(austin_max_idx, austin_max_prcp),
#                xytext=(austin_max_idx, austin_max_prcp + 5),
#                arrowprops=dict(arrowstyle='->', color='m')
#               )


ax[1].plot(seattle.index, seattle['MLY-PRCP-NORMAL'],color='r')
ax[1].plot(seattle.index, seattle['MLY-PRCP-25PCTL'], linestyle='--', color='r')
ax[1].plot(seattle.index, seattle['MLY-PRCP-75PCTL'], linestyle='--', color='r')

annotate_with_marker(seattle, ax[1], loc=(0, -5))

plt.show()
```

```python
fig, ax = plt.subplots(2, 1)

ax[0].plot(austin.index, austin,color='b')
ax[0].plot(austin.index, austin_weather['MLY-PRCP-25PCTL'], linestyle='--', color='b')
ax[0].plot(austin.index, austin_weather['MLY-PRCP-75PCTL'], linestyle='--', color='b')

ax[1].plot(seattle.index, seattle['MLY-PRCP-NORMAL'],color='r')
ax[1].plot(seattle.index, seattle['MLY-PRCP-25PCTL'], linestyle='--', color='r')
ax[1].plot(seattle.index, seattle['MLY-PRCP-75PCTL'], linestyle='--', color='r')

plt.show()
```

```python
medals = pd.read_csv('datasets/medals_by_country_2016.csv', index_col=0)
medals
```

```python
fig, ax = plt.subplots()


ax.bar(medals.index, medals['Gold'], label='Gold')

ax.set_xticks(range(len(medals.index)))
ax.set_xticklabels(medals.index, rotation=90)

ax.set_ylabel('Number of medals')

ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()


ax.bar(medals.index, medals['Gold'], label='Gold')
ax.bar(medals.index, medals['Silver'],bottom=medals['Gold'] ,label='Silver')
ax.bar(medals.index, medals['Bronze'], bottom=medals['Gold'] + medals['Silver'],label='Bronze')


ax.set_xticks(range(len(medals.index)))
ax.set_xticklabels(medals.index, rotation=90)

ax.set_ylabel('Number of medals')

ax.legend()
plt.show()
```

```python
import numpy as np

fig, ax = plt.subplots()

n = len(medals)
x = np.arange(n)

bar_width = 0.25


ax.bar(x - bar_width, medals['Gold'], width=bar_width, label='Gold')
ax.bar(x, medals['Silver'], width=bar_width,label='Silver')
ax.bar(x + bar_width,  medals['Bronze'], width=bar_width, label='Bronze')


ax.set_xticks(range(len(medals.index)))
ax.set_xticklabels(medals.index, rotation=90)

ax.set_ylabel('Number of medals')

ax.legend()
plt.show()
```

```python
summer = pd.read_csv('datasets/summer2016.csv')
summer
```

```python
mens_rowing = summer.loc[(summer['Sport'] == 'Rowing') & (summer['Sex'] == 'M')]
mens_rowing.head()
```

```python
mens_gymnasctics = summer.loc[(summer['Sport'] == 'Gymnastics') & (summer['Sex'] == 'M')]
mens_gymnasctics.head()
```

```python
fig, ax = plt.subplots()


ax.hist(mens_rowing['Weight'], bins=5, label='Rowing', histtype='step')
ax.hist(mens_gymnasctics['Weight'], bins=5, label='Gymnastics', histtype='step')

ax.set_xlabel('Weight (kg)')
ax.set_ylabel('# of observation')

ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()


ax.bar('Rowing',mens_rowing['Height'].mean(), yerr=mens_rowing['Height'].std(),label='Rowing', )
ax.bar('Gymnastics',mens_gymnasctics['Height'].mean(), yerr=mens_gymnasctics['Height'].std(),label='Gymnastics')

ax.set_xlabel('Heihgt (sm)')
ax.set_ylabel('Avg')

ax.legend()
plt.show()
```
