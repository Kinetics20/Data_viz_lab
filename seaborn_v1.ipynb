```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/schoolimprovement2010grants.csv')
```

```python
df.head()
```

```python
df['Award_Amount'].plot.hist()

plt.show()
```

```python
sns.histplot(df['Award_Amount'], bins=20)

plt.show()
```

```python


sns.displot(
    df['Award_Amount'],

    kind='kde',
    fill=True,
    rug=True
)

plt.show()
```

```python
df = pd.read_csv('datasets/insurance_premiums.csv')
df.head()
```

```python
sns.regplot(x='insurance_losses', y='premiums', data=df)

plt.show()
```

```python
# Faceting

val = sns.lmplot(x='insurance_losses', y='premiums', data=df, col='Region')
print(type(val))
plt.show()
```
