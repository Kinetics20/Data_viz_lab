```python
g = sns.JointGrid(
    x='hum',
    y='total_rentals',
    data=bike,
    xlim=(0.1, 1.0)
)

g.plot(sns.regplot, sns.histplot)

plt.show()
```

```python
g = sns.JointGrid(
    x='hum',
    y='total_rentals',
    data=bike,
    xlim=(0.1, 1.0)
)

g.plot_joint(sns.kdeplot)
g.plot_marginals(sns.kdeplot, fill=True)

plt.show()
```

```python
# sns.set_palette('inferno')
sns.jointplot(
        x='hum',
    y='total_rentals',
    data=bike,
    kind='hex',
    cmap='inferno'
)

plt.show()
```

```python
(sns.jointplot(
    x='hum',
    y='total_rentals',
    data=bike.query("total_rentals < 4000"),
    kind='scatter'

)).plot_joint(sns.kdeplot)

plt.show()
```
