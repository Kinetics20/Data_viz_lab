```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
fig, ax = plt.subplots()
plt.show()
```

```python
seattle_weather = pd.read_csv('datasets/seattle_weather.csv')
seattle_weather.head()
```

```python
seattle_weather.columns
```

```python
seattle_weather.loc[:10,'NAME': 'MLY-CLDD-BASE55']
```

```python
seattle_weather['MLY-TAVG-NORMAL']
```

```python
seattle_avg_t = seattle_weather.groupby('DATE')['MLY-TAVG-NORMAL'].mean()
```

```python
fig, ax = plt.subplots()
ax.plot(seattle_avg_t)

plt.show()
```

```python
austin_weather = pd.read_csv('datasets/austin_weather.csv')
```

```python
seattle_avg_prcp = seattle_weather.groupby('DATE')['MLY-PRCP-NORMAL'].mean()
austin_avg_prcp = austin_weather.groupby('DATE')['MLY-PRCP-NORMAL'].mean()
```

```python
fig, ax = plt.subplots()

ax.plot(seattle_avg_prcp.index, seattle_avg_prcp.values)
ax.plot(austin_avg_prcp.index, austin_avg_prcp.values)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(seattle_avg_prcp.index, seattle_avg_prcp.values, color='b', marker='o', linestyle='--')
ax.plot(austin_avg_prcp.index, austin_avg_prcp.values, color='r', marker='v', linestyle='-.')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(seattle_avg_prcp.index, seattle_avg_prcp.values, color='b', marker='o', linestyle='--', label='seattle')
ax.plot(austin_avg_prcp.index, austin_avg_prcp.values, color='r', marker='v', linestyle='-.', label='austin')

ax.set_xlabel('Time (months)')
ax.set_ylabel('Precipitation (inches)')
ax.set_title('Average precipitation for Seattle and Austin')

ax.legend()
plt.show()
```

## Small multipless

```python
fig, ((ax0, ax1),(ax2, ax3)) = plt.subplots(nrows=2, ncols=2)

ax2.set_title('Something')

plt.show()
```

```python
austin = austin_weather.groupby('DATE')[['MLY-PRCP-NORMAL', 'MLY-TAVG-NORMAL']].mean()
seattle = seattle_weather.groupby('DATE')[['MLY-PRCP-NORMAL', 'MLY-TAVG-NORMAL']].mean()
```

```python
austin.info()
```

```python
fig, ax = plt.subplots(2, 2, sharey='col')

ax[0, 0].plot(austin.index, austin['MLY-PRCP-NORMAL'])
ax[0, 1].plot(austin.index, austin['MLY-TAVG-NORMAL'])

ax[1, 0].plot(seattle.index, seattle['MLY-PRCP-NORMAL'])
ax[1, 1].plot(seattle.index, seattle['MLY-PRCP-NORMAL'])



plt.show()
```
