```python
austin_weather['MONTH'] = pd.to_datetime(austin_weather['DATE'], format='%m').dt.strftime('%b')
austin_weather['MONTH'].head(n=13)
```

```python
months = {
    1: "Jan",
    2: "Feb",
    3: "Mar",
    4: "Apr",
    5: "May",
    6: "Jun",
    7: "Jul",
    8: "Aug",
    9: "Sep",
    10: "Oct",
    11: "Nov",
    12: "Dec"
}

seattle_weather['MONTH'] = seattle_weather['DATE'].apply(lambda x: months[x])
seattle_weather['MONTH'].head()
```

```python
austin = austin_weather.groupby('MONTH', sort=False)[['MLY-PRCP-NORMAL', 'MLY-PRCP-25PCTL', 'MLY-PRCP-75PCTL']].mean()
seattle = seattle_weather.groupby('MONTH', sort=False)[['MLY-PRCP-NORMAL', 'MLY-PRCP-25PCTL', 'MLY-PRCP-75PCTL']].mean()
```

```python
def annotate_with_marker(df, ax, title='Max Value', loc=(0, 0),**arrowprops):
    max_idx_, max_val_ = df['MLY-PRCP-NORMAL'].idxmax(), df['MLY-PRCP-NORMAL'].max()

    props = dict(arrowstyle='->', color='r') | arrowprops

    ax.scatter(max_idx_, max_val_, color='r', marker='o')
    ax.annotate(title,
               xy=(max_idx_, max_val_),
               xytext= (max_idx_ , max_val_ + loc[1]),
               arrowprops=props
               )
```

```python
fig, ax = plt.subplots(2, 1, sharey=True)

ax[0].plot(austin.index, austin['MLY-PRCP-NORMAL'], color='b')
ax[0].plot(austin.index, austin['MLY-PRCP-25PCTL'], linestyle='--', color='b')
ax[0].plot(austin.index, austin['MLY-PRCP-75PCTL'], linestyle='--', color='b')


ax[0].axvline(austin['MLY-PRCP-NORMAL'].idxmax(), color='purple', linestyle='--', linewidth=0.5)
ax[0].axhline(austin['MLY-PRCP-NORMAL'].max(), color='purple', linestyle='--', linewidth=0.5)

annotate_with_marker(austin, ax[0], loc=(0, 5))


ax[1].plot(seattle.index, seattle['MLY-PRCP-NORMAL'], color='r')
ax[1].plot(seattle.index, seattle['MLY-PRCP-25PCTL'], linestyle='--', color='r')
ax[1].plot(seattle.index, seattle['MLY-PRCP-75PCTL'], linestyle='--', color='r')

ax[1].axvline(seattle['MLY-PRCP-NORMAL'].idxmax(), color='purple', linestyle='--', linewidth=0.5)
ax[1].axhline(seattle['MLY-PRCP-NORMAL'].max(), color='purple', linestyle='--', linewidth=0.5)

annotate_with_marker(seattle, ax[1], loc=(0, -5))

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(austin.index, austin, color='b')
ax.plot(austin.index, austin_weather['MLY-PRCP-25PCTL'], linestyle='--', color='b')
ax.plot(austin.index, austin_weather['MLY-PRCP-75PCTL'], linestyle='--', color='b')



ax.plot(seattle.index, seattle['MLY-PRCP-NORMAL'], color='r')
ax.plot(seattle.index, seattle['MLY-PRCP-25PCTL'], linestyle='--', color='r')
ax.plot(seattle.index, seattle['MLY-PRCP-75PCTL'], linestyle='--', color='r')

# plt.style.use('default')
plt.show()
```

```python
medals = pd.read_csv('datasets/medals_by_country_2016.csv', index_col=0)
medals.head()
```

```python
fig, ax = plt.subplots()

ax.bar(medals.index, medals['Gold'], label='Gold Medals')

ax.set_xticks(range(len(medals.index)))
ax.set_xticklabels(medals.index, rotation=90)

ax.set_ylabel('Number of Medals')
ax.set_xlabel('Countries')

plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar(medals.index, medals['Gold'], label='Gold Medals')
ax.bar(medals.index, medals['Silver'], bottom=medals['Gold'],label='Silver Medals')
ax.bar(medals.index, medals['Bronze'], bottom=medals['Gold'] + medals['Silver'],label='Bronze Medals')

ax.set_xticks(range(len(medals.index)))
ax.set_xticklabels(medals.index, rotation=90)

ax.set_ylabel('Number of Medals')
ax.set_xlabel('Countries')

plt.legend()
plt.show()
```

```python
import numpy as np

fig, ax = plt.subplots()

n = len(medals)
x = np.arange(n)

bar_width = 0.25

ax.bar(x - bar_width, medals['Gold'], width=bar_width, label='Gold Medals')
ax.bar(x, medals['Silver'], width=bar_width, label='Silver Medals')
ax.bar(x + bar_width, medals['Bronze'], width=bar_width, label='Bronze Medals')

ax.set_xticks(range(len(medals.index)))
ax.set_xticklabels(medals.index, rotation=90)

ax.set_ylabel('Number of Medals')
ax.set_xlabel('Countries')

plt.legend()
plt.show()
```

```python
summer = pd.read_csv('datasets/summer2016.csv')
summer.head()
```

```python
mens_rowing = summer.loc[(summer['Sport'] == 'Rowing') & (summer['Sex'] == 'M')]
mens_rowing.head()
```

```python
mens_gymnastics = summer.loc[(summer['Sport'] == 'Gymnastics') & (summer['Sex'] == 'M')]
mens_gymnastics.head()
```

```python
fig, ax = plt.subplots()

ax.hist(mens_rowing['Weight'], bins=5, label='Rowing', histtype='step')
ax.hist(mens_gymnastics['Weight'], bins=5, label='Gymnastics', histtype='step')

ax.set_xlabel('Weight (kg)')
ax.set_ylabel('# of observation')

plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar('Rowing',mens_rowing['Height'].mean(), yerr=mens_rowing['Height'].std(),label='Rowing')
ax.bar('Gymnastics',mens_gymnastics['Height'].mean(),  yerr=mens_gymnastics['Height'].std(),label='Gymnastics' )

ax.set_xlabel('Height (cm)')
ax.set_ylabel('Avg')

plt.legend()
plt.show()
```
