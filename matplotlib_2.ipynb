```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
fig, ax = plt.subplots()
plt.show()
```

```python
seattle_weather = pd.read_csv('datasets/seattle_weather.csv')
seattle_weather.head()
```

```python
seattle_weather.columns
```

```python
seattle_weather.loc[:10,'NAME': 'MLY-CLDD-BASE55']
```

```python
seattle_weather['MLY-TAVG-NORMAL']
```

```python
seattle_avg_t = seattle_weather.groupby('DATE')['MLY-TAVG-NORMAL'].mean()
```

```python
fig, ax = plt.subplots()
ax.plot(seattle_avg_t)

plt.show()
```

```python
austin_weather = pd.read_csv('datasets/austin_weather.csv')
```

```python
seattle_avg_prcp = seattle_weather.groupby('DATE')['MLY-PRCP-NORMAL'].mean()
austin_avg_prcp = austin_weather.groupby('DATE')['MLY-PRCP-NORMAL'].mean()
```

```python
fig, ax = plt.subplots()

ax.plot(seattle_avg_prcp.index, seattle_avg_prcp.values)
ax.plot(austin_avg_prcp.index, austin_avg_prcp.values)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(seattle_avg_prcp.index, seattle_avg_prcp.values, color='b', marker='o', linestyle='--')
ax.plot(austin_avg_prcp.index, austin_avg_prcp.values, color='r', marker='v', linestyle='-.')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(seattle_avg_prcp.index, seattle_avg_prcp.values, color='b', marker='o', linestyle='--', label='seattle')
ax.plot(austin_avg_prcp.index, austin_avg_prcp.values, color='r', marker='v', linestyle='-.', label='austin')

ax.set_xlabel('Time (months)')
ax.set_ylabel('Precipitation (inches)')
ax.set_title('Average precipitation for Seattle and Austin')

ax.legend()
plt.show()
```

```python
austin = austin_weather.groupby('DATE')[['MLY-PRCP-NORMAL', 'MLY-TAVG-NORMAL']].mean()
seattle = seattle_weather.groupby('DATE')[['MLY-PRCP-NORMAL', 'MLY-TAVG-NORMAL']].mean()
```

```python
austin.info()
```

```python
fig, ax = plt.subplots(2, 2, sharey='col')

ax[0, 0].plot(austin.index, austin['MLY-PRCP-NORMAL'])
ax[0, 1].plot(austin.index, austin['MLY-TAVG-NORMAL'])

ax[1, 0].plot(seattle.index, seattle['MLY-PRCP-NORMAL'])
ax[1, 1].plot(seattle.index, seattle['MLY-PRCP-NORMAL'])



plt.show()
```

```python
austin_weather['MONTH'] = pd.to_datetime(austin_weather['DATE'], format='%m').dt.strftime('%b')
austin_weather['MONTH'].head(n=13)
```

```python
months = {
    1: "Jan",
    2: "Feb",
    3: "Mar",
    4: "Apr",
    5: "May",
    6: "Jun",
    7: "Jul",
    8: "Aug",
    9: "Sep",
    10: "Oct",
    11: "Nov",
    12: "Dec"
}

seattle_weather['MONTH'] = seattle_weather['DATE'].apply(lambda x: months[x])
seattle_weather['MONTH'].head()
```

```python
austin = austin_weather.groupby('MONTH', sort=False)[['MLY-PRCP-NORMAL', 'MLY-PRCP-25PCTL', 'MLY-PRCP-75PCTL']].mean()
seattle = seattle_weather.groupby('MONTH', sort=False)[['MLY-PRCP-NORMAL', 'MLY-PRCP-25PCTL', 'MLY-PRCP-75PCTL']].mean()
```

```python
fig, ax = plt.subplots(2, 1, sharey=True)

ax[0].plot(austin.index, austin['MLY-PRCP-NORMAL'], color='b')
ax[0].plot(austin.index, austin['MLY-PRCP-25PCTL'], linestyle='--', color='b')
ax[0].plot(austin.index, austin['MLY-PRCP-75PCTL'], linestyle='--', color='b')

ax[1].plot(seattle.index, seattle['MLY-PRCP-NORMAL'], color='r')
ax[1].plot(seattle.index, seattle['MLY-PRCP-25PCTL'], linestyle='--', color='r')
ax[1].plot(seattle.index, seattle['MLY-PRCP-75PCTL'], linestyle='--', color='r')

# plt.style.use('default')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(austin.index, austin, color='b')
ax.plot(austin.index, austin_weather['MLY-PRCP-25PCTL'], linestyle='--', color='b')
ax.plot(austin.index, austin_weather['MLY-PRCP-75PCTL'], linestyle='--', color='b')

ax.plot(seattle.index, seattle['MLY-PRCP-NORMAL'], color='r')
ax.plot(seattle.index, seattle['MLY-PRCP-25PCTL'], linestyle='--', color='r')
ax.plot(seattle.index, seattle['MLY-PRCP-75PCTL'], linestyle='--', color='r')

# plt.style.use('default')
plt.show()
```
