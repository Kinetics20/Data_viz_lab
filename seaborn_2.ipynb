```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/schoolimprovement2010grants.csv')
df.head()
```

```python
df['Award_Amount'].plot.hist()

plt.style.use('ggplot')
plt.show()
```

```python
sns.histplot(df['Award_Amount'], bins=20, kde=True)
# sns.set()
# plt.style.use('tableau-colorblind10')
plt.show()
```

```python
import seaborn as sns
sns.reset_defaults()
```

```python

sns.displot(
    df['Award_Amount'],
    kind='kde',
    fill=True,
    rug=True
)

plt.show()
```

```python
df = pd.read_csv('datasets/insurance_premiums.csv')
df.head()
```

```python
sns.regplot(x='insurance_losses', y='premiums', data=df)

plt.show()
```

```python


sns.lmplot(x='insurance_losses', y='premiums', data=df, col='Region')

plt.show()
```
