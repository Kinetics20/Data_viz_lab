```python
sns.set_style('ticks')
sns.color_palette('rocket', 20)

fig, ax= plt.subplots()

sns.histplot(df['fmr_1'], ax=ax, kde=True)

ax.set(
    xlabel='1 Bedroom Fair Market Rent',
    title='US Rent',
    xlim=(300, 1500),
    ylabel='Frequency'

)

median_ = df['fmr_1'].median()
mean_ = df['fmr_1'].mean()

ax.axvline(x=median_, color='r', linestyle='--', label='Median')
ax.axvline(x=mean_, color='g', linestyle='-', label='Mean')
ax.axvspan(min(mean_, median_), max(mean_, median_), color='steelblue', alpha=.3, label='Mean-Median Range')

plt.legend()


sns.despine()
plt.show()
```

```python
fig, ax= plt.subplots()

sns.histplot(df['fmr_1'], ax=ax)
sns.histplot(df['fmr_2'], ax=ax)

sns.despine()
plt.show()
```

```python
fig, (ax0, ax1)= plt.subplots(1, 2, sharey=True, sharex=True, figsize=(12, 5))

sns.histplot(df['fmr_1'], ax=ax0)
ax0.set(xlabel='1 Bedroom')

sns.histplot(df['fmr_2'], ax=ax1)
ax1.set(xlabel='2 Bedroom')

sns.despine()
plt.tight_layout()
plt.show()
```

## Categorical data

```python
school = pd.read_csv('datasets/schoolimprovement2010grants.csv')
school.head()
```

```python
sns.stripplot(
    x='Award_Amount',
    y='Model Selected',
    data=school,
    jitter=True
)

plt.show()
```

```python

sns.set_palette('bright')

sns.boxplot(
    y='Award_Amount',
    x='Model Selected',
    hue='Model Selected',
    data=school

)

plt.show()
```

```python
sns.set_palette('bright')

sns.countplot(
    y='Model Selected',
    hue='Region',
    data=school
)

plt.show()
```

```python
sns.swarmplot(
    x='Award_Amount',
    y='Model Selected',
    data=school,
    size=2
)

plt.show()
```

```python
sns.violinplot(
    x='Award_Amount',
    y='Model Selected',
    data=school,

)

plt.show()
```

```python
sns.boxenplot(
    x='Award_Amount',
    y='Model Selected',
    data=school,
    hue='Model Selected'
)
plt.show()
```

```python
sns.barplot(
    x='Award_Amount',
    y='Model Selected',
    data=school,
    hue='Model Selected'
)
plt.show()
```

```python
sns.pointplot(
    x='Award_Amount',
    y='Model Selected',
    data=school,
    hue='Model Selected'
)
plt.show()
```
