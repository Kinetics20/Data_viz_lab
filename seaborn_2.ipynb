```python
sns.set_style('ticks')
sns.color_palette('rocket', 20)

fig, ax= plt.subplots()

sns.histplot(df['fmr_1'], ax=ax, kde=True)

ax.set(
    xlabel='1 Bedroom Fair Market Rent',
    title='US Rent',
    xlim=(300, 1500),
    ylabel='Frequency'

)

median_ = df['fmr_1'].median()
mean_ = df['fmr_1'].mean()

ax.axvline(x=median_, color='r', linestyle='--', label='Median')
ax.axvline(x=mean_, color='g', linestyle='-', label='Mean')
ax.axvspan(min(mean_, median_), max(mean_, median_), color='steelblue', alpha=.3, label='Mean-Median Range')

plt.legend()


sns.despine()
plt.show()
```

## Faceting

```python
df = pd.read_csv('datasets/college_datav3.csv')
df.head()
```

```python
g = sns.FacetGrid(
    df,
    col='Degree_Type',
    col_order=['Graduate', 'Bachelors', 'Associates', 'Certificate']
)

g.map(sns.pointplot, 'SAT_AVG_ALL')
plt.show()
```

```python
sns.catplot(
    x='Tuition',
    data=df,
    kind='box',
    col='Degree_Type',
)

plt.show()
```

```python
g = sns.FacetGrid(df, col='Degree_Type')
g.map(plt.scatter, 'UG', 'PCTPELL')

plt.show()
```

```python
sns.lmplot(x='UG', y='PCTPELL', data=df, col='Degree_Type')

plt.show()
```

```python
sns.lmplot(
    x='Tuition',
    y='SAT_AVG_ALL',
    data=df,
    row='REGION',
    col='HIGHDEG'
)

plt.show()
```

```python
df_1 = pd.read_csv('datasets/insurance_premiums.csv')
df_1.head()
```

```python
g = sns.PairGrid(df_1, vars=['fatal_collisions', 'premiums'])
g2 = g.map(sns.scatterplot)

print(type(g2))
plt.show()
```

```python
g = sns.PairGrid(df_1, vars=['fatal_collisions', 'premiums'])
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)

plt.show()
```

```python
sns.pairplot(df_1, vars=['fatal_collisions', 'premiums'], kind='reg', diag_kind='kde')

plt.show()
```

```python
sns.pairplot(
    data=df_1,
    x_vars=['fatal_collisions_speeding', 'fatal_collisions_alc'],
    y_vars=['insurance_losses', 'premiums'],
    kind='scatter',
    hue='Region',
    palette='husl'
)

plt.show()
```
